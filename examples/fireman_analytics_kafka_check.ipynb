```python
from kafka import KafkaConsumer
from influxdb import InfluxDBClient
import json
```

# 1. Check if measurements are written to InfluxDB

```python
client = InfluxDBClient(host='localhost', port=8086, database='fireman')
client.query("select * from data limit 10")
```

# 2. Check if Collector & Classifier are writing measurements & predictions to Kafka stream

```python
consumer_data = KafkaConsumer(
        'spam_data',
        bootstrap_servers='localhost:19092',
        value_deserializer=lambda value: json.loads(value),
    )
```

```python
consumer_predictions = KafkaConsumer(
        'spam_predictions',
        bootstrap_servers='localhost:19092',
        value_deserializer=lambda value: json.loads(value),
    )
```

```python
for message in consumer_data:
    print(message)
```

```python
for message in consumer_predictions:
    print(message)
```
